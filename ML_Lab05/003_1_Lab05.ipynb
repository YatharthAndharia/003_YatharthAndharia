```python
import numpy as np
```

```python
X = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
Y = np.array([[56], 
                    [81], 
                    [119], 
                    [22], 
                    [103]], dtype='float32')
```

```python
#Normalization
mu = np.mean(X, 0)
sigma = np.std(X, 0)
X = (X-mu) / sigma
X = np.hstack((np.ones((Y.size,1)),X))
print(X)
```

```python
#Obtaining random weight values
rng = np.random.default_rng(3)
w = rng.random((1, 4))
print(w)
```

```python
def mse(mat1, mat2):
    diff = mat1 - mat2
    return np.sum(diff * diff) / diff.size
```

```python
#row wise sigma(xi*wi)
def model(x,w):
    return x @ w.T
```
