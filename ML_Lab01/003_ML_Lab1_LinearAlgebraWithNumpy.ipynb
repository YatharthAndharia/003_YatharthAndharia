```python
import numpy as np
import pandas as pd

matrix32 = np.array([[4, 5], [1, 2], [3, 6]])
matrix23 = np.array([[1, 4, 3], [2, 4, 5]])

print('3 x 2 matrix : \n', matrix32)
print('\n 2 x 3 matrix : \n', matrix23)
```

```python
random_matrix = np.random.rand(3, 4)
print(random_matrix)
```

```python
# matrix multiplication
matmul = np.dot(matrix32, matrix23)
print(matmul)
```

```python
# elemet wise matrix multiplication
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[4,5,6],[1,2,3]])
result=np.multiply(a,b)
print('Element wise Matrix Multiplication : \n', result)
```

```python
mat=np.matrix([[1,2,3],[4,5,6],[7,8,9]])
mean = np.mean(mat)
print(mean)
```
