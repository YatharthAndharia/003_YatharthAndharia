# Try SVM classifier on MNIST dataset, compare the preformance of linear, polynomial and RBF kernels.

```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
print(digits.target)
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.3 ,random_state = 3)
```

# Linear SVM Classifier

```python
lsvm = svm.SVC(kernel='linear')
```

```python
lsvm.fit(X_train, y_train)
```

```python
l_y_pred = lsvm.predict(X_test)
```

```python
print("Accuracy of Linear SVM:", metrics.accuracy_score(y_test, l_y_pred))
```

# Polynominal SVM Classifier

```python
poly_svm = svm.SVC(kernel='poly')
poly_svm.fit(X_train, y_train)
poly_y_pred = poly_svm.predict(X_test)
```

```python
print("Accuracy of Polynomial SVM:", metrics.accuracy_score(y_test, poly_y_pred))
```

# RBF SVM Classifier

```python
rbf_svm = svm.SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
rbf_y_pred = rbf_svm.predict(X_test)
```

```python
print("Accuracy of RBF SVM:", metrics.accuracy_score(y_test, rbf_y_pred))
```
