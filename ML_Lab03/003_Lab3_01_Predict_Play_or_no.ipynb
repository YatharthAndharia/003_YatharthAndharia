```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
dataset = pd.read_csv("Dataset1.csv")
print(dataset)
```

```python
label_encoder = preprocessing.LabelEncoder()
Y_rows=None
for data_heading in dataset:
  if data_heading!="Play":
    print(f"\n\nHeading :- {data_heading}")
    dummy = pd.get_dummies(dataset[data_heading])
    dataset = dataset.drop([data_heading],axis=1)
    dataset = pd.concat([dataset,dummy],axis=1)
  else:
    #print(dataset[data_heading])
    Y_rows = label_encoder.fit_transform(dataset[data_heading])
    print(Y_rows)
    dataset = dataset.drop([data_heading],axis=1)
```

```python
print(dataset)
print (Y_rows)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(dataset, Y_rows, test_size = 0.30, random_state = 3)
```

```python
print(X_train)
print()
print(X_test)
print()
print(Y_train)
print(Y_test)
```

```python
model = MultinomialNB()
model.fit(X_train, Y_train)
Y_predicted = model.predict(X_test)
print(X_test)
print(Y_predicted)
```

```python
from sklearn import metrics

print(f"Accuracy is :- {metrics.accuracy_score(Y_test, Y_predicted)}")

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score


precision = precision_score(Y_test, Y_predicted)
recall = recall_score(Y_test, Y_predicted)


print(f"precision :- {precision}")
print(f"recall :- {recall}")
```
